# Data merge(데이터 병합)

```python
import pandas as pd
import numpy as np
```

```python
df_a = pd.DataFrame({'key': ['a','b','c','d','e'], 'num_a': [1,2,3,4,5]})
df_b = pd.DataFrame({'key': ['a','b','d','f','g'], 'num_b': [11,15,35,45,55]})
df_c = pd.DataFrame({'key': ['f','g','h','i','j'], 'num_a': [6,7,8,9,0]})
```

```python
df_a
```

```python
df_b
```

```python
df_c
```

```python
pd.concat([df_a, df_b])
```

```python
pd.concat([df_a, df_c])
```

```python
pd.concat([df_a, df_b, df_c])# 여러개 파일 한 줄로 합치기 가능. #없는 부분은 null값
```

```python
pd.concat([df_a, df_b, df_c], axis = 1) # axis = 1 가로로 붙이기
```

```python
df_a.merge(df_b, how = 'inner') # 기본이 inner(how생략가능) 공통된 키값만 보여줌
```

```python
df_a.merge(df_b, how = 'outer') # 모든걸 합치며 없는 값은 null
```

```python
df_a.merge(df_b, how = 'left') # 왼쪽에 있는 키값을 기반으로 합치기
```

```python
df_a.merge(df_b, how = 'right') # 오른쪽에 있는 키값을 기반으로 합치기
```

```python
df_a.merge(df_b, on = 'key' ,how = 'right') 
# 합치는 기준은 pandas가 알아서 찾아줌
# 지정도 가능함 >> on = 'key'
```

```python
df_a = pd.DataFrame({'key': ['a','b','c','d','e'], 'id': ['q','w','e','r','t'] ,'num_a': [1,2,3,4,5]})
df_b = pd.DataFrame({'key': ['a','b','d','f','g'], 'id': ['r','t','z','x','y'] ,'num_b': [11,15,35,45,55]})
```

```python
df_a
```

```python
df_b
```

```python
df_a.merge(df_b, on = 'key')
```

```python
df_a.merge(df_b, on = 'id')
```

```python
# 합치는 3가지 방법
# 1. concat - 위아래로 합치기. 원하면 옆으로도 합칠 수 있음(axis)
## 키값이 아닌 단순히 옆으로 붙이는 방법
# 2. merge -- 데이터 프레임에 존재하는 컬럼을 기준으로 합치기
# 3. join -- 인덱스를 기준으로 합치기
```

```python
salary_1 = pd.read_csv('salary_1.csv')
salary_2 = pd.read_csv('salary_2.csv')
# 동일한 컬럼의 두 데이터
```

```python
salary_1.head()
```

```python
# 머신러닝 데이터를 만드는 목적의 경우 groupby를 사용할 일이 더 많음
# 피봇 테이블은 데이터를 원하는 형태로 간결하게 보기위해서 등 사용
```

```python
sales_df = pd.DataFrame({'company': ['a','a','a','a','b','b','b','b'],
             'quarter': ['q1','q2','q3','q4','q1','q2','q3','q4'],
             'sales': [111,222,333,444,555,666,777,888]})
```

```python
sales_df
```

```python
# 피봇테이블은 연산
# 피봇은 데이터를 뿌려줌
sales_temp = pd.pivot(sales_df, index = 'company', columns = 'quarter', values = 'sales')
```

```python
sales_temp.columns
```

```python
sales_temp.columns = sales_temp.columns.rename('')
```

```python
sales_temp
```

```python
new_sales_df = sales_temp.reset_index()
```

```python
new_sales_df
```

```python
# 옆으로 퍼진 데이터를 다시 원래 형태로 길쭉하게 재배치
# melt >> 피벗 해제
# 1. id_vars 처음 기준이 되는 컬럼(기준점)
# 2. value_vars 하나의 컬럼이 될 데이터 리스트(해체 대상)
# 3. var_name 헤더(해체 대상) 이름표 -- var 는 variable (변수)
# 4. value_name 내용물의 이름표
# 5. .sort_values 로 정렬
pd.melt(new_sales_df, id_vars = 'company', value_vars = ['q1','q2','q3','q4'], var_name ='quarter' ,value_name = 'sales').sort_values('company')
```

# 주성분 분석 (PCA), 복습 (Recap)

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA()
```

```python
pca.fit(salary_df)
```

```python
salary_df
```

```python
# pca.transform(salary_df) -- numpy array 형태로 보여줌
pd.DataFrame(pca.transform(salary_df))
```

```python
pca = PCA(2)
```

```python
pd.DataFrame(pca.fit_transform(salary_df), columns = ['PC1','PC2'])
```

```python
pd.DataFrame(pca.transform(salary_df))
```

```python
pca = PCA(2)
```

```python
pd.DataFrame(pca.fit_transform(salary_df), columns = ['PC1','PC2'])
```

```python
(pca.explained_variance_ratio_)
```

```python
(pca.explained_variance_ratio_).sum()
```

```python
salary_df.corr()
```

```python
# PCA
# 차원축소
# 데이터의 정보를 최대한 간직하는 방법으로 변수를 줄임
# 노이즈가 껴있는걸 줄이고
# 다중공산성 문제 해결

# PCA의 단점?
# 어떤 변수들이 포함되어 있는지 알 수 없음
# 변수의 특성을 설명할 길이 없다
# 그래프를 그릴 수 있지만 x축과 y축이 무엇인지 말할 수 없음
# 데이터 있는 그래도의 결과가 좋은지, PCA의 결과가 더 좋은지
# PCA가 예측력이 좋을 수 있지만 항상 좋은 것은 아니다
# 예측력도 증요하지만 설명을 할 수 있어야함
# 설명이 중요하다면 PCA가 좋은 선택이 아닐 수 있다.
```
