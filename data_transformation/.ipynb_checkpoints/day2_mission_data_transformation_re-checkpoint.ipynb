### 자료형 확인 & 자료형을 숫자 형태로 변환

```python
car_df.info()
```

```python
# mileage 컬럼 데이터에서 kmpl 문자 제거하고 숫자 부분만 저장
car_df['mileage'] = car_df['mileage'].str.replace('kmpl','')
```

```python
# engine 컬럼 데이터에서 CC 문자 제거하고 숫자 부분만 저장
car_df['engine'] = car_df['engine'].str.replace('CC','')
```

```python
# engine 컬럼 데이터에서 CC 문자 제거하고 숫자 부분만 저장
car_df['max_power'] = car_df['max_power'].str.replace('bhp','')
```

```python
car_df['mileage'].head(10)
```

```python
# mileage 컬럼에 km/kg 로 단위가 다른 컬럼 발견
# 다시 제거하고 숫자 부분만 저장
car_df['mileage'] = car_df['mileage'].str.replace(' km/kg','')
```

```python
car_df.info()
```

```python
car_df['engine'] = pd.to_numeric(car_df['engine'])
```

```python
car_df['max_power'] = car_df['max_power'].str.replace(' ','')
```

```python
car_df['max_power'] = pd.to_numeric(car_df['max_power'])
```

```python
car_df['mileage'] = pd.to_numeric(car_df['mileage'])
```

```python
# 원-핫 인코딩 적용 -- 첫번째 범주 제외함

car_data = pd.get_dummies(car_df, drop_first = True)
```

```python
from sklearn.preprocessing import RobustScaler
```

```python
rs = RobustScaler()
```

```python
rs.fit(car_data) # fit >> 학습을 시킴
```

```python
rs_df = pd.DataFrame(rs.transform(car_data), columns = car_data.columns)
```

```python
rs_df.describe()
```

```python
rs.fit_transform(rs_df)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA()
```

```python
pca.fit(rs_df)
```

```python
rs_df
```

```python
pd.DataFrame(pca.transform(rs_df))
```

```python
for i in range(2, 11):
    pca = PCA(i)
    pca.fit(rs_df)
    print(i, round(pca.explained_variance_ratio_.sum(), 2))
```

```python
pca_result = PCA()
```

```python
explained_var = pca.explained_variance_ratio_
```

```python
explained_var[:5]
```

```python
cumulative_var = np.cumsum(explained_var)
```

```python
cumulative_var[:5]
```

```python
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(cumulative_var)+1), cumulative_var, marker='o', linestyle='-', color='purple', label='Cumulative Explained Variance')

# y = 0.80 선
plt.axhline(y=0.80, color='red', linestyle='--', label='80% Threshold')

# y = 0.90 선
plt.axhline(y=0.90, color='dodgerblue', linestyle='--', label='90% Threshold')

# 80%를 넘는 첫 번째 컴포넌트 표시
n_components_80 = np.argmax(cumulative_var >= 0.80) + 1
plt.axvline(x=n_components_80, color='red', linestyle='--', label=f'{n_components_80} Components')

# 90%를 넘는 첫 번째 컴포넌트 표시
n_components_90 = np.argmax(cumulative_var >= 0.90) + 1
plt.axvline(x=n_components_90, color='dodgerblue', linestyle='--', label=f'{n_components_90} Components')

# 누적 그레프 
plt.xlabel('Number of PC')
plt.ylabel('Cumulative Explained Variance Ratio')
plt.grid(True)
plt.xticks(range(1, len(cumulative_var)+1))
plt.ylim(0.4, 1.05)
plt.tight_layout()
plt.show()
```
