## 구매금액의 z-score 계산하기
### z-score로 구매 패턴 확인

```python
# z-score 계산을 위한 실습용 데이터

temp = pd.DataFrame({'a': [10,20,30,20,10,200], 'b': [100,300,200,150,250,200], 'c': [10, 500, 20, 250, 25, 200]})
```

```python
temp
```

```python
# temp 데이터의 컬럼별 평균 계산

temp.mean()
```

```python
# temp 데이터의 컬럼별 표준편차 계산

temp.std()
```

```python
# 각 데이터에 대해 a 컬럼의 z-score 계산
# (data - mean) / std

(temp['a'] - 48.33) / 74.67
```

```python
# 각 데이터에 대해 b 컬럼의 z-score 계산 

(temp['b'] - 200) / 70.71
```

```python
# 각 데이터에 대해 c 컬럼의 z-score 계산 

(temp['c'] - 167.50) / 192.50
```

```python
# trans_date_trans_time 컬럼에서 시간 값을 추출 >> datetime 자료형 이용

pd.to_datetime(cc_df['trans_date_trans_time']).dt.hour
```

```python
# hour 컬럼에 저장

cc_df['hour'] = pd.to_datetime(cc_df['trans_date_trans_time']).dt.hour
```

```python
cc_df.head()

# 평균적인 결제 시간을 기준으로 z-score 계산 가능
# 하지만 컴퓨터 기준으로 0시과 23시의 gap이 커짐
# 따라서 시간을 그룹 단위로 분류
```

```python
# 결제시간을 morning, afternoon, night, evening으로 분류하기 위해 함수 정의

def hour_func(x):
    if (x >= 6) & (x < 12):
        return 'morning'
    elif (x >= 12) & (x < 18):
        return 'afternoon'
    elif (x >= 18) & (x < 23):
        return 'night'
    else:
        return 'evening'
```

```python
# hour 컬럼에 hour_func 함수를 적용

cc_df['hour'].apply(hour_func)
```

```python
# hour_cat 컬럼에 저장

cc_df['hour_cat'] = cc_df['hour'].apply(hour_func)
```

```python
rs_cc
```

```python
pd.DataFrame(pca.transform(rs_cc))
```

```python
explained_var = pca.explained_variance_ratio_
```

```python
cumulative_var = np.cumsum(explained_var)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(cumulative_var)+1), cumulative_var, marker='o', linestyle='-', color='purple', label='Cumulative Explained Variance')

# y = 0.80 선
plt.axhline(y=0.80, color='red', linestyle='--', label='80% Threshold')

# y = 0.90 선
plt.axhline(y=0.90, color='dodgerblue', linestyle='--', label='90% Threshold')

# 80%를 넘는 첫 번째 컴포넌트 표시
n_components_80 = np.argmax(cumulative_var >= 0.80) + 1
plt.axvline(x=n_components_80, color='red', linestyle='--', label=f'{n_components_80} Components')

# 90%를 넘는 첫 번째 컴포넌트 표시
n_components_90 = np.argmax(cumulative_var >= 0.90) + 1
plt.axvline(x=n_components_90, color='dodgerblue', linestyle='--', label=f'{n_components_90} Components')

# 누적 그레프 
plt.xlabel('Number of PC')
plt.ylabel('Cumulative Explained Variance Ratio')
plt.grid(True)
plt.xticks(range(1, len(cumulative_var)+1))
plt.ylim(0.75, 1.05)
plt.tight_layout()
plt.show()
```
